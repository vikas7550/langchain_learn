```python
from dotenv import load_dotenv

load_dotenv()
```

## No memory

```python
from langchain_openrouter import ChatOpenRouter

model = ChatOpenRouter(
    model="nvidia/nemotron-3-super-120b-a12b:free",
    temperature=0.1,
)
```

```python
from langchain.agents import create_agent


agent = create_agent(
    model=model
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
from pprint import pprint

pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## Memory

```python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    model=model,
    checkpointer=InMemorySaver(),  
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
```

```python
question = HumanMessage(content="my current favourite colour is blue")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
```

```python
question = HumanMessage(content="what is my favourite colour?")

from langchain.messages import SystemMessage

system_message = SystemMessage(content="You are a helpful assistant that remembers the user's favourite colour. and you only answer the color name. ")

response = agent.invoke(
    {"messages": [system_message, question]},
    config,  
)

pprint(response)
```
